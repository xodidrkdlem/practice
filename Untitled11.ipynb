```python
#if문 정리
pocket = ['paper', 'cellphone']  #포켓이라는 리스트안에 'paper', 'cellphone'이라는 객체가 있다.
card = True                      #불자료형으로 card는 항상 True이다.
if "money" in pocket:            #만약에 pocket안에 money라는 객체가 있다면 
  print("택시를 타고 가라")      #택시를 타고 가라를 출력해라

else:                            #위에 if가 거짓일 경우
  if card:                       #'money'가 pocket안에 없을 시 만약 card라면? //위에서 card는 항상 True라고 알려줌.
    print("택시를 타고 가라")    #택시를 타고 가라를 출력
  else:                          #money'가 pocket안에 없을 시 만약 card도 없다면 
    print("걸어가라")            #걸어가라를 출력해라
```

```python
#while 문

# while (조건문):
#   수행할 문장
#   수행할 문장 2
#   수행할 문장 3
#   ....

#예제 1

treehit  = 0                                     #treehit은 0번
while treehit < 10:                              #treehit이 10보다 작을때 동안
  treehit = treehit + 1                          #treehit이 반복되는 동안(즉 한번 돌 때마다 하나씩 증가한다) ==[treehit += 1]
  print("나무를 %d번 찍었습니다." %treehit)      #나무를 N번 찍었습니다 출력
  if treehit == 10:                              #만약에 treehit이 10이면 
    print("나무 넘어갑니다.")                    #나무 넘어갑니다를 출력해라
```

```python

#while문 예제

coffee = 10                                           #coffee 는 10
money = 300                                           #money 는 300
while money:                                          #money가 참인 동안: (money라는 변수안에 객체가 있으므로 True)
  print("돈을 받았으니 커피를 줍니다.")               #문자열 출력
  coffee -= 1                                         #coffee 숫자가 하나씩 줄어든다.(while문을 돌때마다)
  print("남은 커피의 양은 %d개입니다" % coffee)       #문자열 출력
  if coffee == 0:                                     # 만약에 커피의 개수가 0이라면
    print("커피가 떨어졌습니다. 판매를 중지합니다.")  # 출력(커피가 떨어졌습니다. 판매를 중지합니다)
    break                                             #빠져나오는 것.

```

```python
#coffee 자판기 예제

coffee = 10                                                       #coffee = 10
while True:                                                       #참인 동안
  money = int(input("돈을 넣어주세요: "))                         #money 안에 입력값을 받는다
  if money ==300:                                                 #만약 입력으로 300을 받는다면
    print("커피를 줍니다.")                                       #커피를 줍니다, 남은 커피의 양을 출력한다.
    print("남은 커피의 양은 %d개입니다."% coffee )
    coffee -= 1                                                   #money가 300일때 커피는 하나씩 줄어든다.
  elif money >300:                                                #money가 300이상일때
    print("거스름돈 %d와 커피를 줍니다." % (money- 300))          #거스름돈(입력값-300)과 커피를 줍니다 출력
    coffee -= 1                                                   #커피가 하나씩 줄어든다.
    print("남은 커피의 양은 %d개입니다."% coffee )                # 남은 커피양은 %d개입니다.
  else:                                                           #위의 말이 모두 거짓이였을 때
    print("돈을 다시 돌려주고 커피를 주지않습니다.")              #출력
    print("남은 커피의 양은 %d개입니다."% coffee )                #출력
  if coffee == 0:                                                 #만약에 커피가 0이라면.
    print("커피가 소진되었습니다. 판매를 중지합니다.")            #커피가 소진되었습니다. 판매를 중지합시다.
    break                                                         #if == 0이라면 빠져나온다는 뜻.
    
```

```python
#for문
#예제1.

test = ['one', 'two', 'Three']   #test 라는 리스트안에 one, two, three 라는 객체가 있다.
for i in test:                   #i 에 차례대로 test안에 있는 객체 대입
  print(i)


a = [(1,2),(3,4),(5,6)]          #a라는 리스트안에 튜플형태의 (1,2), (3,4), (5,6) 가 있다.
for (first, last) in a:          # (first, last) 에 a가 순차적으로 대입
  print(first + last)            #(first +last)한 후에 순차적으로 출력 


  
```

```python
#for문 응용

marks = [90,25,67,45,80]                             #marks안에 90,25,67,45,80라는 객체가 있다.
number = 0                                           #number = 0 이라고 선언
for mark in marks:                                   #mark안에 순서대로 marks를 대입한다.
  number += 1                                        #for문이 돌때마다 number가 하나씩 늘어난다.
  if mark >= 60:
    print("%번 학생은 합격입니다." % number)
  else:
    print("%번 학생은 불합격입니다."% number)

```

```python
#for 문과 continue 문
marks = [90,25,67,48,80]

number  = 0                                   #number = 초기화
for mark in marks:                            #mark안에 있는 객체가 mark를 향해
  number += 1                                 #number는 하나씩 증가한다.
  if mark >= 60: continue                     #만약에 mark가 60보다 크거나 같으면 continue(== 한마디로 스킵.//다시 for문 첫번째로)
  print("%d번 학생 축하드린다" %number)       #출력(번 학생 축하드린다.)

```

```python
#for문과 같이 사용하는 range 함수

 #range(10)    ----->  10미만까지  0~9까지
 #range(0,10)    -----> 0~9까지  

add = 0            #----> add는 0

for i in range(1,11):         #i 안에 1부터 10까지 차례대로 대입
   add = add + i              # add(0) + 1 , add(1) + 2, add(3) + 3 .......
   print(add)                 # add 출력 

#예제 2
#for문과 range 함수를 사용하여 1부터 100까지 더해보자.

first = 0

for i in range(1,101):
  first = first + i
  print(first) 


```

```python
#인덱싱, len()함수를 활용한 for문

marks = [90,25,67,45,80]

for number in range(len(marks)):
  if marks[number] < 60: continue
  print("%d번 학생 축하드립니다. 합격입니다." %(number +1))
```

```python
#for문을 활용한 구구단 만들기

for i in range(2,10):               #i에는 2~9까지
  for j in range(1,10):             #j에는 1~9까지
    print(i*j, end = " ")           #출력해라(i가 2일때 j는 1~9까지 하나 하나 대입후 for을 벗어나고, 다시 i가 3일때 ..,)

  print("\n")                      #jin range(1,10) 이 끝나면 줄바꾸기
```

```python
#리스트 내포사용하기

a = [1,2,3,4]            #a라는 리스트 안에 1,2,3,4 가있다.
result = []              # result라는 리스트 만듬
for num in a:            #num이라는 변수안에 a리스트안에 있는 객체 하나씩 넣기
  result.append(3*num)   #result라는 리스트에 3*1,3*2,3*3,3*4 추가하기 append 리스트에 추가하는 함수

print(result)            #result 출력하기
```
